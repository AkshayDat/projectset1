# Removing Outliers and skewness from the dataset

```python
from scipy.stats import zscore
z=np.abs(zscore(sal))
z
```

```python
#removing the outliers
```

```python
threshold=3
```

```python
#printing all the values which are greater than the threshold value
print(np.where(z>3))
```

```python
z[7][2]#checking wether the result is correct or not
```

```python
z[188][3]#checking wether the result is correct or not
```

```python
sal_new=sal[(z<3).all(axis=1)]#removing the outliers
```

```python
#checking the skewness after removing the outliers
sal_new.skew()
```

```python
colllist=sal_new.columns.values
for i in range(0,len(colllist)):
    if sal_new.skew().loc[colllist[i]]>0.55:
        sal_new[colllist[i]]=np.sqrt(sal_new[colllist[i]])
```

```python
sal_new.skew()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
max_score=0
final_rstate=0

for i in range(1,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=i)
    lm=LinearRegression() 
    lm.fit(x_train,y_train)
    lm.score(x_train,y_train)
    pred=lm.predict(x_test)
    r_score=r2_score(y_test,pred)
    print("R2 SCORE is equal to :",r_score,"at random state",i)
    
    if r_score > max_score:
        max_score=r_score
        final_rstate=i
        
        
    
    
print("The maximum r2_score is ",max_score, "at random state",final_rstate) 
```
